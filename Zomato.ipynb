```python
df_country=pd.read_excel('/content/Country-Code[1].xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##pie chart Top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct="'%1.2f%%")
```

```python
#observation:Zomato maximum records or transactions are from India after that United states and then united kingdom.
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
##observation
#1.when ratings b\w 4.5 to 4.9 ---> Excellent
#2.when ratings b\w 4 to 4.5 ---> Very good
#3.when ratings b\w 3.5 to 3.9 ---> good
#4.when ratings b\w 2.5 to 3.4   ---> Average
#5.when ratings b\w 1.8 to 2.4 ---> poor
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=['blue','Red','Orange','Yellow','Green','Green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','Red','Orange','Yellow','Green','Green'])
```
